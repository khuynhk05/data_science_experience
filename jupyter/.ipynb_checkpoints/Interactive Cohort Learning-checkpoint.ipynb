```python
# @hidden_cell
%matplotlib inline
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
from IPython.display import clear_output
#from datetime import datetime

sns.set(style='white')
mpl.rcParams['lines.linewidth'] = 2
ds= None
df= None
cohort_select= [None, None]
ds_select=None 

# Load the data frame from the dataset, print the first 6throw
def load_data_frame():
  global df, ds, cohort_select, ds_select
  df = pd.read_csv("../datasets/%s" % ds)
  print "Rows x Columns : %s" % str(df.shape)

def cohort_period(df):
  df['CohortPeriod'] = np.arange(len(df)) + 1
  return df

def plot(user_retention):
  user_retention['Average'] = user_retention.mean(axis=1)

  user_retention.plot(figsize=(10,8))
  plt.title('Cohorts: User Retention')
  plt.xticks(np.arange(1, 12.1, 1))
  plt.xlim(1, 12)
  plt.ylabel('% of Cohort Purchasing');

  plt.figure(figsize=(12, 8))
  plt.title('Cohorts: User Retention')
  sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.0%');

def run(event):
  global df, ds, cohort_select, ds_select

  #Group by cohort = month
  format_column = "Formatted%s" % cohort_select[0].value
  df.set_index('UserId', inplace=True)
  df['CohortGroup'] = df.groupby(level=0)[format_column].min()
  df.reset_index(inplace=True)
  #Total
  df['TotalCharges'] = (df['TotalCharges'].replace( '[\$,)]','', regex=True ).replace( '[(]','-',   regex=True ).astype(float))

  #group by cohort
  grouped = df.groupby(['CohortGroup', format_column])

  cohorts = grouped.agg({'UserId': pd.Series.nunique,
                       'OrderId': pd.Series.nunique,
                       'TotalCharges': np.sum})
  cohorts.rename(columns={'UserId': 'TotalUsers', 'OrderId': 'TotalOrders'}, inplace=True)
    
  #Label the group
  cohorts = cohorts.groupby(level=0).apply(cohort_period)

  cohorts.reset_index(inplace=True)
  cohorts.set_index(['CohortGroup','CohortPeriod'],inplace=True)
  cohort_group_size = cohorts['TotalUsers'].groupby(level=0).first()

  #calculate retention percentage
  user_retention = cohorts['TotalUsers'].unstack(0).divide(cohort_group_size, axis=1)
  plot(user_retention)
    
def cohort_group_selected(change):
  global df, ds, cohort_select, ds_select
  if change['type'] == 'change' and change['name'] == 'value' and change['owner'].value != "Select a Date Column as Cohort Group":
    cohort_c = cohort_select[0].value
    df[cohort_c] = pd.to_datetime(df[cohort_c])
    format_column = "Formatted%s" % cohort_c
    df[format_column] = df[cohort_c].apply(lambda x: x.strftime('%Y-%m'))
    clear_output()
     
def display_cohort_selector():
  global df, ds, cohort_select, ds_select
  clear_output(wait=True)
  load_data_frame()
  headers = [x for x in df.columns]
  index_headers = headers[:]
  try: 
    headers.insert(0, "Select a Date Column as Cohort Group")
    if cohort_select[0] is None:
      cohort_select[0] = widgets.Dropdown(options=headers)
      cohort_select[0].observe(cohort_group_selected)
      display(cohort_select[0])
    
      btn = widgets.Button(description="Run")
      btn.on_click(run)
      display(btn)
    else:
      cohort_select[0].options = headers
    cohort_select[0].value = "Select a Date Column as Cohort Group"
  except Exception, e:
    print str(e)

# Observe dataset input changes
def dataset_selected(change):
  global df, ds, cohort_select, ds_select
  if change['type'] == 'change' and change['name'] == 'value' and change['new'] != 'Select Dataset':
    clear_output(wait=True)
    ds = change['new']
    display_cohort_selector()

def main():
  global df, ds, cohort_select, ds_select
  try:     
    files = [f for f in os.listdir('../datasets') if str(f).endswith("csv")]
    files.insert(0, "Select Dataset")    
    ds_select = widgets.Dropdown( options=files)
    ds_select.observe(dataset_selected)
    display(ds_select)
  except Exception, e:
    print str(e)
main()
```
